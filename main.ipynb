```python
import pandas as pd
import json
```

```python
result_path = "results_2025-11-08T00-17-34.641690.json"

with open(result_path, 'r') as f: 
    data = json.load(f)
    result_dict = data['results']
    df = pd.DataFrame.from_dict(result_dict, orient='index')

df
```

```python
df.columns = [column_name.replace('none', '' )for column_name in df.columns]
df.columns = [column_name.replace(',', '' )for column_name in df.columns]
```
